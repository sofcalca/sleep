```python
SQRD_ACC_X=pd.DataFrame()
SQRD_ACC_Y=pd.DataFrame()
SQRD_ACC_Z=pd.DataFrame()
for col in acc_spectrum.columns:
    if col[4]=="X":
        SQRD_ACC_X[col[6:]]=acc_spectrum[col]**2
    elif col[4]=="Y":
        SQRD_ACC_Y[col[6:]]=acc_spectrum[col]**2
    elif col[4]=="Z":
        SQRD_ACC_Z[col[6:]]=acc_spectrum[col]**2
MSE_ACC=pd.DataFrame()
MSE_ACC= np.sqrt(SQRD_ACC_X + SQRD_ACC_Y + SQRD_ACC_Z)
```

```python
MSE_ACC
```

```python
def count_by_zero(array):
    counts = 0
    for i in range(0, len(array)-1):
        if np.sign(array[i]) != np.sign(array[i+1]):
            counts+=1
    return counts
```

```python
centered_MSE = (MSE_ACC.T-MSE_ACC.mean(axis=1)).T 
```

```python
centered_ACC_through_0=centered_MSE.apply(count_by_zero, axis=1)
```

```python
centered_ACC_through_0.name="centered_ACC_through_0"
```

```python
MSE_ACC_quantiles = MSE_ACC.quantile(np.linspace(0,1,11), axis=1).T
cols_int = ["{}_quantile_MSE_ACC".format(k) for k in range(0,110,10)]
MSE_ACC_quantiles.columns = [name for name in cols_int]

MSE_ACC_means = pd.Series(MSE_ACC.mean(axis=1), name = "mean_MSE_ACC")
                          
MSE_ACC_var = pd.Series(MSE_ACC.var(axis=1), name= "var_MSE_ACC")

MSE_ACC_skew = pd.Series(MSE_ACC.skew(axis=1), name = "skew_MSE_ACC")
                         
MSE_ACC_kurt = pd.Series(MSE_ACC.kurtosis(axis=1), name="kurt_MSE_ACC")
```

```python
ACC_stats = pd.concat(
    [centered_ACC_through_0, MSE_ACC_quantiles, MSE_ACC_means, MSE_ACC_var, MSE_ACC_skew, MSE_ACC_kurt],
    axis=1
)
```
